```python
import pandas as pd

df = pd.read_csv("Adjusted_Model.csv")

df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
revenue_by_park = df.groupby("Park_Name")["Adjusted_Revenue"].sum()
```

```python
plt.figure(figsize=(10,6))
revenue_by_park.plot(kind="bar")
plt.show()
```

```python
revenue_by_park.plot(kind="bar")

plt.title("Total Revenue by Park")
plt.xlabel("Park")
plt.ylabel("Revenue")
plt.tight_layout()

plt.savefig("Revenue_by_Park.png", dpi = 300)

plt.show()
```
